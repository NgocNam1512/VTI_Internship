Ở đây, em dùng các yếu tố có độ liên quan >= 0.05 để dự đoán giá nhà.


```
X = data.loc[:, ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'sqft_above', 'sqft_basement']]
y = data.loc[:, ['price']]
```

Phân chia dữ liệu thành tập train và test

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Sử dụng thuật toán Linear Regression để dự đoán giá

```
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
test_pre = lr.predict(X_test)
```

Các thông số đánh giá:

```
lr.score(X_test, y_test)  #r2 score
```

```
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("Mean absolute Error: {:.2f}".format(mean_absolute_error(y_test, test_pre)))
print("Mean squared error: {:.2f}".format(mean_squared_error(y_test, test_pre)))
print("R2 score: {:.2f}".format(r2_score(y_test, test_pre)))
```
